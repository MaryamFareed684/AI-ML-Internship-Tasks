```python
# Pandas Library for importing dataset
import pandas as pd
```

```python
#loadung iris dataset
df=pd.read_csv("iris.csv")
```

```python
#importing libraries for visualizing data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#relationship between sepal length and petal length
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=df
)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
#relationship between sepal width and petal width
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='sepal_width',
    y='petal_width',
    hue='species',
    data=df
)
plt.title('Sepal width vs Petal width')
plt.show()
```

```python
#Plotting using histogram. It shows:
#How values are distributed
#Frequency of values
#Shape of data
#Range of values

df.drop('species', axis=1).hist(
    figsize=(10,8),
    bins=15
)
plt.suptitle('Feature Distributions')
plt.show()
```

```python
#Outliers are datapoints that are usually low or high as compare to rest of data
#Box plot to identify outliers in datset
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df.drop('species', axis=1)
)
plt.title('Box Plot for Outlier Detection')
plt.show()
```

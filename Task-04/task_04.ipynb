```python
#imports the pipeline function from Hugging Face Transformers.
from transformers import pipeline

#Creating a pipeline object called generator to generate text based on your inputs.
generator = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",#flan-t5-base is a general-purpose instruction-following model by Google.
    framework="pt", #This tells the pipeline to use PyTorch as the backend.
    max_new_tokens=200  #This limits how long the generated answer can be.
)
```

```python
#This defines a Python function called get_health_response.
def get_health_response(q):
    prompt = f"""
Act as a friendly medical assistant.             
Answer general health questions clearly.
Do NOT give diagnosis or prescriptions.

Question: {q}
Answer:
"""
    # This tells the model how to behave.
    return generator(prompt)[0]["generated_text"]
```

```python

while True:
    user = input("You: ")
    if user.lower() in ["exit", "quit"]:
        print("Bot: Take care! ðŸ˜Š")
        break
    print("Bot:", get_health_response(user))
```
